```python
crop_df = pd.read_csv("D:\\12\\ML Thesis\\November\\Dataset_N.csv")
crop_df.head()
```

Exploring the data

```python
print("Information \n", crop_df.info)
print("\n")
print("Keys \n", crop_df.keys())
print("\n")
print("Dtypes \n", crop_df.dtypes)
```

Encoding into categories

```python
crop_df = crop_df.join(pd.get_dummies(crop_df['Area']))
crop_df = crop_df.join(pd.get_dummies(crop_df['Year']))
crop_df = crop_df.join(pd.get_dummies(crop_df['Item']))
crop_df
```

Dropped unnecessary data

```python
crop_df=crop_df.drop('Area', axis=1)
crop_df=crop_df.drop('Year', axis=1)
crop_df=crop_df.drop('Item', axis=1)
crop_df=crop_df.drop('Production', axis=1)
crop_df
```

Processing the data

```python
# Create x, where x the 'scores' column's values as floats
x = crop_df[['Yield']].values.astype(int)
x
# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(x)

# Run the normalizer on the dataframe
#df_normalized = pd.DataFrame(x_scaled)
x_scaled

crop_df['yield'] = x_scaled
crop_df
```

```python
crop_df.head()
```

Split the data into test and train

```python
y = crop_df['Yield']
X = crop_df.drop('Yield', axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
from sklearn import svm
classifier = svm.SVC(kernel='linear',gamma='auto',C=2)
classifier.fit(X_train,y_train)

y_predict = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

### **Generate the model**

Selecting kernel

```python
cls = svm.SVC(kernel="linear")
```

Training the model

```python
cls.fit(X_train,y_train)
```

Predict the response

```python
crop_df = cls.predict(X_test)
```

Printing accuracy

```python
print("Accuracy:", metrics.accuracy_score(y_test,y_pred=crop_df))
```

Printing precision score

```python
print("Precision:", metrics.precision_score(y_test,y_pred=crop_df))
```
