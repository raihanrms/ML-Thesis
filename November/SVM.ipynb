```python
import pandas as pd
crop_df=pd.read_csv("D:\\12\\ML Thesis\\November\\Dataset_N.csv")
crop_df
```

```python
crop_df.keys()
```

```python
crop_df = crop_df.join(pd.get_dummies(crop_df['Area']))
crop_df = crop_df.join(pd.get_dummies(crop_df['Year']))
crop_df = crop_df.join(pd.get_dummies(crop_df['Item']))
crop_df
```

```python
crop_df=crop_df.drop('Area', axis=1)
crop_df=crop_df.drop('Year', axis=1)
crop_df=crop_df.drop('Item', axis=1)
crop_df=crop_df.drop('Production', axis=1)
crop_df
```

```python
from sklearn import preprocessing
```

```python
# Create x, where x the 'scores' column's values as floats
x = crop_df[['Area Harvested']].values.astype(int)
x
# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(x)

# Run the normalizer on the dataframe
#df_normalized = pd.DataFrame(x_scaled)
x_scaled

crop_df['Area Harvested'] = x_scaled
crop_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
b = crop_df['Yield']
a = crop_df.drop('Yield', axis = 1)

a_train, a_test, b_train, b_test = train_test_split(a, b, test_size = 0.25, random_state = 0)

print(a_train)
print(a_test)
print(b_train)
print(b_test)
```

```python
b_train
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel="poly")
classifier.fit(a_train,b_train)
```

```python
from sklearn.metrics import accuracy_score
b_pred = classifier.predict(a_test)
accuracy_score(b_test, b_pred)
```
